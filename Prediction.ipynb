```python
import tensorflow as tf
from tensorflow.keras.models import load_model
import pickle
import numpy as np
import pandas as pd
```

```python
scaler = pickle.load(open('/content/drive/MyDrive/ANN Practice/scaler.pkl','rb'))
ohe = pickle.load(open('/content/drive/MyDrive/ANN Practice/onehot_encoder.pkl','rb'))
le = pickle.load(open('/content/drive/MyDrive/ANN Practice/label_encoder.pkl','rb'))
model = load_model('/content/drive/MyDrive/ANN Practice/model.h5')
```

```python
input_data = {
  'CreditScore': 700,
  'Geography':'France',
  'Gender':'Male',
  'Age':40,
  'Tenure':3,
  'Balance':60000,
  'NumOfProducts':2,
  'HasCrCard':1,
  'IsActiveMember':1,
  'EstimatedSalary':50000
}
```

```python
df = pd.DataFrame(input_data,index=[0])
```

```python
df['Gender'] = le.transform(df['Gender'])
```

```python
encoded_geo = ohe.transform(df[['Geography']])
encoded_df = pd.DataFrame(encoded_geo.toarray(),columns=ohe.get_feature_names_out(['Geography']))
df = pd.concat([df,encoded_df],axis=1)
df.drop('Geography',axis=1,inplace=True)
df
```

```python
input_scaled = scaler.transform(df)
```

```python
prediction = model.predict(input_scaled)
prediction
```
